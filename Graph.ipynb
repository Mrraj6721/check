```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
X=[1,8,0,16,8,4,10,2]
Y=[4,7,1,3,5,7,0,12]

df = pd.DataFrame({'X': X, 'Y': Y})  
```

```python
plt.bar(X,Y)
plt.show()
```

```python
plt.plot(X,Y)
plt.show()
```

```python
plt.hist(X,bins=10)
plt.show()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
